```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders

# ======================= #
#  CONFIGURATION SECTION  #
# ======================= #

sender_email = "omnimmalwar13@gmail.com"
app_password = "orgm twqo qzua iqsm"          # Gmail App Password
receiver_email = "omnimmalwar852004@gmail.com"

subject = "Project Report and Jupyter Notebook"
body = """
Hello,

Please find attached the project report and the Jupyter Notebook containing the full analysis.

Regards,
Om Nimmalwar
"""

# Files to attach
attachments = [
    # "/mnt/data/jupyter_report.pdf",
    "miniPorjectNoteBook.ipynb",
    "miniPorjectNote.pdf",
    "All Documentations And Reports/FinalReport.pdf"
]

# ======================= #
#   EMAIL SENDING LOGIC   #
# ======================= #

# Create message object
msg = MIMEMultipart()
msg["From"] = sender_email
msg["To"] = receiver_email
msg["Subject"] = subject

# Add email body
msg.attach(MIMEText(body, "plain"))

# Attach all files
for file_path in attachments:
    part = MIMEBase("application", "octet-stream")
    with open(file_path, "rb") as file:
        part.set_payload(file.read())

    encoders.encode_base64(part)
    part.add_header(
        "Content-Disposition",
        f"attachment; filename={file_path.split('/')[-1]}"
    )
    msg.attach(part)

# ======================= #
#     SEND EMAIL (SMTP)   #
# ======================= #

with smtplib.SMTP("smtp.gmail.com", 587) as server:
    server.starttls()
    server.login(sender_email, app_password)
    server.sendmail(sender_email, receiver_email, msg.as_string())

print("Email sent successfully!")
```
