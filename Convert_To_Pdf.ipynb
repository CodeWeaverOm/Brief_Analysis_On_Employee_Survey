```python
import json
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.pagesizes import A4

# Load notebook
notebook_path = "miniPorjectNoteBook.ipynb"
with open(notebook_path, "r", encoding="utf-8") as f:
    nb = json.load(f)

# PDF output path
pdf_path = "miniPorjectNote.pdf"

styles = getSampleStyleSheet()
story = []

# Extract markdown + code
for cell in nb.get("cells", []):
    if cell["cell_type"] == "markdown":
        text = "".join(cell["source"])
        story.append(Paragraph(text.replace("\n", "<br/>"), styles["BodyText"]))
        story.append(Spacer(1, 12))

    elif cell["cell_type"] == "code":
        code = "".join(cell["source"])
        code_html = f"<font face='Courier'>{code.replace('<', '&lt;').replace('>', '&gt;').replace('\n', '<br/>')}</font>"
        story.append(Paragraph(code_html, styles["Code"]))
        story.append(Spacer(1, 12))

# Build PDF
doc = SimpleDocTemplate(pdf_path, pagesize=A4)
doc.build(story)

print("PDF created:", pdf_path)
```
