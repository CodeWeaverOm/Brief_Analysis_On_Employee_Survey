```python
#Importing Required Packages
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Importinh Data Set
df = pd.read_csv("All Data Sets/employee_survey.csv")
df.head()
```

**As per Above Information There Is No Null Values In DataFrame So We Move Further Toward Our Next Step That Is finding Outliers In Important Columns**

```python
plt.figure(figsize=(12,10))
plt.subplot(3, 3, 1)
sns.boxplot(y=df['Age'])
plt.title("Outlier in Age")

# 2nd subplot
plt.subplot(3, 3, 2)
sns.boxplot(y=df['SleepHours'])
plt.title("Outlier in SleepHours")

# 3rd subplot
plt.subplot(3, 3, 3)
sns.boxplot(y=df['Experience'])
plt.title("Outlier in Experience")

# 4th subplot
plt.subplot(3, 3, 4)
sns.boxplot(y=df['PhysicalActivityHours'])
plt.title("Outlier in PhysicalActivityHours")

plt.tight_layout()
plt.show()
```
